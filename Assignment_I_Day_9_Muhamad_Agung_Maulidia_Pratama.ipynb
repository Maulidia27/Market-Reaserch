Susunan Pipeline dalam membuat suatu Model Machine learning


1.   Import Data
      > Sebelum kita mulai masuk ketahap pembuatan model machine learning, kita   harus mengubah data agar dapat dimasukkan kedalam model machine learning kita. Import data adalah proses memuat dan membaca data ke notebook. Hal yang harus diperhatikan dalam import data adalah format data dan lokasi penyimpanan file dataset. Format data yang dapat di-import ke notebook bervariasi seperti .csv, .xlsx, .json, dan sebagainya, tentunya setelah import data, harus kita pahami isi data tersebut. [Klik Reference](https://ilmudatapy.com/import-dan-export-data-di-python/#:~:text=Import%20data%20adalah%20proses%20memuat,%2C%20.json%2C%20dan%20sebagainya.)

2. Feature Engineering & Selection
      > Dalam data preparation termasuk proses data cleaning, menangani data outliers dan pemilihan feature. Jika Feature Engineering & Selection dilakuakn sebelum train test split ada kemungkinan nilai akurasi yang didapat overfitting, sedangkan jika dilakukan setelah train test split, data train dan data test dapat dilakukan normalisasi,standarisasi, polynomial, or dll sehingga dapat menunjukan nilai akurasi yang sebenar-benarnya
[teks link](https://stackoverflow.com/questions/56308116/should-feature-selection-be-done-before-train-test-split-or-after)
3.   Train_Test Split
      > Secara umum, kita akan membagi data kita menjadi 3 bagian : Training, Testing, Validation. Kita telah melatih model dengan data training, mengevaluasinya dengan data validation dan yang terakhir mengujinya dengan data testing. agar model yang kita buat dapat menggeneralisasi dengan baik pada data yang belum pernah dilatih, dengan kata lain, memprediksi hasil yang akurat dari data yang baru.
4.   Data Imputation
      > Setelah kita mengimport data, untuk memastikan data yang kita miliki harus kita imputasi agar data kita memiliki kualitas yang baik untuk diolah atau memperkirakan nilai pada missing data dengan cara mendapatkan pola dari data yang memiliki fitur lengkap. Beberapa metode imputasi yang populer adalah: Mean, Median, Modus dan klasterisasi. [Reference](https://www.dqlab.id/digital-transformation-pahami-teknik-pengolahan-ini-dalam-industri-data)
. 
5. Oversampling
      > Oversampling digunakan ketika jumlah data tidak mencukupi. Mencoba menyeimbangkan dataset dengan meningkatkan ukuran sampel langka. Daripada membuang sampel berlimpah, sampel langka baru dihasilkan dengan menggunakan mis. SMOTE (Sintetis Minoritas Sampling Teknik). Jika Cross-Validation diterapkan setelah oversampling, pada dasarnya yang dilakukan adalah overfitting model dengan hasil bootstrap buatan tertentu.Itulah sebabnya Cross-Validation harus selalu dilakukan sebelum oversampling data, seperti halnya bagaimana pemilihan fitur harus dilaksanakan. Hanya dengan melakukan resampling data berulang kali, keacakan dapat dimasukkan ke dalam dataset untuk memastikan bahwa tidak akan ada masalah overfitting.
6.   Modeling
proses membuat representasi visual dari keseluruhan sistem informasi atau bagian-bagiannya untuk mengkomunikasikan koneksi antara titik data dan struktur.
Pemilihan model adalah proses pemilihan salah satu model sebagai model akhir yang menjawab masalah. Sehingga tahapan ini hampir menjadi tahapan akhir dalam proses Machine learning
      Setelah kita menentukan metode yang cocok untuk data yang ada, kita perlu mengubah hyperparameter untuk mendapatkan performa terbaik dari model. Mengubah nilai hyperparameter saat kita menjalankan algoritma ML akan memberikan hasil atau performa model yang berbeda. Proses menemukan performa terbaik model dengan pengaturan hyperparameter yang berbeda ini juga disebut model selection. (https://machinelearningmastery.com/a-gentle-introduction-to-model-selection-for-machine-learning/)

7.   Cross-Validation
      > melakukan validasi berulang di mana dataset dibagi menjadi banyak subset (himpunan) data latih & validasi. Setiap iterasi memvalidasi (menguji) satu subset data dengan subset yang tersisa sebagai data latih. Pada cross validation, # subset data adalah jumlah iterasi.Mengapa kita tidak menyeimbangkan data terlebih dahulu lalu spliting data?, tentu hal ini tidak bisa kita lakukan karena akan menimbulkan kebocoran pada data test, model akan memberikan score tinggi pada data test dan seolah olah model tidak overfit, padahal jika kita terapkan pada data yang benar benar baru, model tersebut kemungkinan besar akan overfit.
Sekilas, cara di atas sudah kita anggap benar, tapi cara tersebut masih memiliki kelemahan saat dilakukan cross validation, melakukan balance dataset sebelum cross validation akan menimbulkan kebocoran pada data validation, hal ini kemungkinan akan menyebabkan model overfit.[teks link](https://kiwidamien.github.io/how-to-do-cross-validation-when-upsampling-data.html)
